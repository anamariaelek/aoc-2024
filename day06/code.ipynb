In this example, the guard will visit 41 distinct positions on your map.

Predict the path of the guard. How many distinct positions will the guard visit before leaving the mapped area?

````python
def find_direction(coord, grid):
    """
    Return the direction the guard is facing at the starting position
    """
    direction_map = {
        '^': 'up',
        'v': 'down',
        '<': 'left',
        '>': 'right'
    }
    symbol = grid[coord[0], coord[1]]
    return direction_map.get(symbol)

def make_step(coord, direction, grid):
    """
    Move the guard one step forward in the given direction
    """
    # Calculate new coordinate
    coord_map = {
        'up' : (coord[0] - 1, coord[1]),
        'down' : (coord[0] + 1, coord[1]),
        'left' : (coord[0], coord[1] - 1),
        'right' : (coord[0], coord[1] + 1)
    }
    coord_new = coord_map.get(direction)
    # Check if the new coordinate is outside the grid
    if is_outside(coord_new, grid):
        coord_new = None
    # Check if there is an obstacle in the new coordinate
    elif is_obstacle(coord_new, grid):
        coord_new = coord
        direction = turn_right(direction)
    return coord_new, direction

def is_outside(coord, grid):
    """ 
    Check if the coordinate is outside the grid
    """
    if coord[0] < 0 or coord[0] >= grid.shape[0] or coord[1] < 0 or coord[1] >= grid.shape[1]:
        return True
    else:
        return False

def is_obstacle(coord, grid):
    """
    Check if there is an obstacle in the given coordinate
    """
    if grid[coord[0], coord[1]] == '#':
        return True
    else:
        return False

def turn_right(direction):
    """ 
    Turn the guard to the right
    """
    turns_map = {
        'up' : 'right',
        'down' : 'left',
        'left' : 'up',
        'right' : 'down'
    }
    return turns_map.get(direction)
````

````python
# Load input
with open('input.txt') as f:
    lines = [line.strip() for line in f]

# Parse iput into grid
import numpy as np
grid = np.array([list(line) for line in lines])

# Find starting position
coord = [int(i[0]) for i in np.where(grid == '^')]

# Now move the guard!
direction = find_direction(coord, grid)
step = 0
coords = []
while coord is not None:    
    if coord not in coords:
        coords.append(coord)
    step += 1
    coord, direction = make_step(coord, direction, grid)

print(len(coords))
````

### Part Two

While The Historians begin working around the guard's patrol route, you borrow their fancy device and step outside the lab. From the safety of a supply closet, you time travel through the last few months and record the nightly status of the lab's guard post on the walls of the closet.

To have the lowest chance of creating a time paradox, The Historians would like to know all of the possible positions for such an obstruction. The new obstruction can't be placed at the guard's starting position - the guard is there right now and would notice.

It doesn't really matter what you choose to use as an obstacle so long as you and The Historians can put it into position without the guard noticing. The important thing is having enough options that you can find one that minimizes time paradoxes, and in this example, there are 6 different positions you could choose.

You need to get the guard stuck in a loop by adding a single new obstruction. How many different positions could you choose for this obstruction?

````python
import numpy as np

# Modfy the function to recognize new obstacle
def is_obstacle(coord, grid):
    """
    Check if there is an obstacle in the given coordinate
    """
    if grid[coord[0], coord[1]] in ['#', 'O']:
        return True
    else:
        return False

# Modify the function to detect loops
def make_step(coord, direction, grid, steps, turns, steps_between_turns, steps_since_turn):
    """
    Move the guard one step forward in the given direction, now also keeping track 
    of turns and how many moves are between them, to identify the loops
    """
    # Calculate new coordinate
    coord_map = {
        'down' : (coord[0] + 1, coord[1]),
        'up' : (coord[0] - 1, coord[1]),
        'left' : (coord[0], coord[1] - 1),
        'right' : (coord[0], coord[1] + 1)
    }
    coord_new = coord_map.get(direction)

    # Update step count
    steps += 1
    steps_since_turn += 1

    # Check if the new coordinate is outside the grid
    if is_outside(coord_new, grid):
        coord_new = None

    # Check if there is an obstacle in the new coordinate
    elif is_obstacle(coord_new, grid):
        coord_new = coord
        direction = turn_right(direction)
        turns += 1
        steps_between_turns.append(steps_since_turn)
        steps_since_turn = 0
        
        # Check if guard is stuck in a loop
        if turns > 4:
            for n in range(2, turns // 2):
                # Get number of step between last n turns
                last_ = steps_between_turns[-n:]
                semi_last_ = steps_between_turns[:-n][-n:]
                if last_ == semi_last_:
                    # Throw an error
                    raise ValueError('Guard is stuck in a loop')
    
    return coord_new, direction, steps, turns, steps_between_turns, steps_since_turn
````

````python
# Load input
with open('example.txt') as f:
    lines = [line.strip() for line in f]

# Parse iput into grid
grid = np.array([list(line) for line in lines])

# Which positions are open?
open_positions = np.argwhere(grid == '.').tolist()
count_loops = 0

from tqdm import tqdm

for pos in tqdm(open_positions):

    # Copy the grid
    grid_copy = np.array([list(line) for line in lines])

    # Place the new obstacle
    grid_copy[pos[0], pos[1]] = 'O'

    # Find starting position
    coord = np.argwhere(grid_copy == '^')[0].tolist()

    # Now move the guard!
    direction = find_direction(coord, grid_copy)
    steps = 0
    turns = 0
    steps_since_turn = 0
    steps_between_turns = []
    coords = []
    try:
        while coord is not None:    
            if coord not in coords:
                coords.append(coord)
            steps += 1
            coord, direction, steps, turns, steps_between_turns, steps_since_turn = make_step(
                coord, direction, grid_copy, steps, turns, steps_between_turns, steps_since_turn)
    except ValueError:
        count_loops += 1

print(f'There are {count_loops} positions where to put a new obstacle')
````
